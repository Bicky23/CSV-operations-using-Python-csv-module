```python
''' Python code for the following tasks:
a) Finding total number of rows in the csv file, ignoring the rows where "prediction" is "None".
b) Finding the distinct values in the "prediction" coloumn of csv.
c) Finding the count of each distinct value appearing in the "prediction" column of csv.
d) Submitting the output of above question into csv file. '''


import csv             # import csv module
count_rows = 0         # Stores the number of rows where "Prediction" is "None". 
distinct_values = {}   # Dictionary containing keys and values as unique values and their count for "Prediction" respectively


with open('interview_scent.csv') as f:              # Reading the csv file 
    file = csv.reader(f)
    header = next(file)
    rows = [row for row in file]
    for row in rows:
        if row[4] != 'None':
            count_rows += 1
        else:
            pass 
    for row in rows:
        target = row[4]
        if "," in target:
            for word in target.split(','):
                if word not in distinct_values.keys():
                    distinct_values[word] = 1
                else:
                    distinct_values[word] += 1
        else:
            if target not in distinct_values.keys():
                distinct_values[target] = 1
            else:
                distinct_values[target] += 1
            
with open('Solution.csv', 'w') as f:               # Writing to a new csv file
    writer = csv.writer(f)
    writer.writerows(distinct_values.items())
```
